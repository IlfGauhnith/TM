## Question 1
Armazene os dados acima em um pandas DataFrame, e processe as urls de cada documento utilizando o Tika. Armazene os textos extraídos em uma coluna chamada `text`.

Store the above data in a pandas DataFrame, and process the urls of each document using Tika. Store the extracted texts in a column named `text`.

```
import re

def extract_urls(text):
  text = str(text)
  urls = re.findall(r'(?:http|ftp|https):\/\/(?:[\w_-]+(?:(?:\.[\w_-]+)+))(?:[\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])', text)
  return urls
```

```
import pandas as pd

df = pd.read_excel('feedbackPortugueseDataset.xlsx')
df = df.drop(['unique_id'], axis=1)

df['urls'] = df['Description.New'].apply(extract_urls)

df = df.dropna()
df = df[df['urls'].apply(lambda urls: len(urls) > 0)]
df = df[['Description.New', 'urls']]
df['url-count'] = df['urls'].apply(lambda urls: len(urls))
df
```

```
def retrieve_text_from_url(url):
  try:
    parsed = parser.from_file(url)
    return parsed['content']

  except Exception as e:
    return f"Error extracting text from {url}: {str(e)}"
```

```
df['text'] = df['urls'].apply(lambda urls: [retrieve_text_from_url(url) for url in urls])
df
```

## Question 2
Utilize a biblioteca hashedindex para criar um índice invertido dos documentos analisados.

Dica:


*   A função `textparser.word_tokenize` retorna uma Tupla, adicione o primeiro elemento dela ao indice (e.g., `term[0]`)
*   Otimize o índice setando o parâmetro `min_length=4` da função `textparser.word_tokenize`

Use the hashedindex library to create an inverted index of analyzed documents.

Tip:


* The `textparser.word_tokenize` function returns a Tuple, add its first element to the index (e.g., `term[0]`)
* Optimize the index by setting the `min_length=4` parameter of the `textparser.word_tokenize` function


```
docs = df['text'].explode().tolist()
errors = [doc for doc in docs if 'error extracting text' in doc.lower()]

print(f'unsuccessful requests ratio: {len(errors)/len(docs)}')
print()
print(errors)
```

```
# Excluding text from unsuccessful requests

docs = [doc for doc in docs if 'error extracting text' not in doc.lower()] # I checked the data previously to garantee there's not valid texts containing 'Error extracting text'.
```

```
import hashedindex
from hashedindex import textparser

index = hashedindex.HashedIndex()

for idx, doc in enumerate(docs):
  tokens = list(textparser.word_tokenize(doc, min_length=4))
  for token in tokens:
    index.add_term_occurrence(token[0], f'document_{str(idx)}')
```
